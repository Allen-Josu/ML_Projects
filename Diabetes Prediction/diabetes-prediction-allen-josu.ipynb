## **Importing Dependencies**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

## **Loading the Data**

```python
# change the path with the current path
path = "/kaggle/input/pima-india/diabetes.csv"
data = pd.read_csv(path)
df=data.copy()
```

## **Exploring the Data**

```python
# printing the first 5 rows of the dataset
df.head()
```

```python
# number of rows and columns in the dataset
df.shape
```

```python
# getting the statistical measures of the data
df.describe()
```

```python
df["Outcome"].value_counts()
```

0 --> Non Diabetic

1 --> Diabetic

```python
df.groupby("Outcome").mean()
```

```python
# Saperating the data and labels 
x = df.drop(columns="Outcome",axis = 1)
y = df["Outcome"]
```

## **Data Standarization**

Evey data in the dataset in not normalized so to predict the data, normalization is a must.

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data = scaler.transform(x)
```

```python
x = standardized_data
y = df["Outcome"]
```

## **Train Test Split**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,stratify=y,random_state = 2)
```

```python
print(df.shape,x_test.shape,x_train.shape)
```

## **Training the Model**

```python
classifier = svm.SVC(kernel="linear")
```

```python
# training the support vector Machine Classifier
classifier.fit(x_train,y_train)
```

## **Model Evaluation**

```python
# accuracy score from the training data
x_train_predict = classifier.predict(x_train)
x_train_accuracy = accuracy_score(x_train_predict,y_train)
print(f"The Accuracy score of Training Data is {x_train_accuracy*100}")
```

```python
# accuracy score from the test data
x_test_predict = classifier.predict(x_test)
x_test_accuracy = accuracy_score(x_test_predict,y_test)
print(f"The Accuracy score of Test Data is {x_test_accuracy*100}")
```

## **Make Prediction**

```python
input_data = (7,103,66,32,0,39.1,0.344,31)

input_array  =  np.asarray(input_data)

# reshaping the input_array
input_array_reshaped = input_array.reshape(1,-1)

# standardization
std_input = scaler.transform(input_array_reshaped)

# prediction
prediction = classifier.predict(std_input)

if prediction[0] == 0:
    print("Non-Diabetes")
else:
    print("Diabeties")
```
